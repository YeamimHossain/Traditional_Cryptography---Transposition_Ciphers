<h1 style = 'text-align: center'><b>Keyed Transposition Cipher</b></h1>
<h3>
    <p>
        In keyed transposition cipher, the plaintext is divided into groups of predetermined size, called blocks, and then use a key to permute the characters in each block separately.
    </p>
</h3>

<h3>Generating necessary list of characters</h3>

```python
# Generating Small letters (English alphabets 'a-z')
SL = []
for x in range(ord('a'),ord('z')+1):
    SL.append(chr(x))
print(SL)
```

```python
# Generating Capital letters (English alphabets 'A-Z')
CL = []
for x in range(ord('A'),ord('Z')+1):
    CL.append(chr(x))
print(CL)
```

```python
PT = SL
CT = CL
Modulus = len(PT)
```

<h3>Input Message</h3>

```python
# message = input("Write your message here: ")
message = "Meet me at the garden"
print(f'Message: {message}')
```

<h3>Encrypting Message</h3>

```python
# List of input characters
inlist = list(message)
is_inlist = 0
while is_inlist<len(inlist):
    if inlist[is_inlist] in CT:
        inlist[is_inlist] = inlist[is_inlist].lower()
    is_inlist+=1
print(inlist)
```

```python
# Removing blank spaces from the message 
mtext = []      # mtext is the message text
for x in inlist:
    if x in PT:
        mtext.append(x)
print(mtext)
```

```python
print(len(mtext))
```

```python
EK = list(input("Enter the encryption key: "))
print(EK)
```

```python
len_msg = len(mtext)
while len_msg % len(EK) != 0:
    len_msg += 1
print(len_msg)
```

```python
bogus_req = len_msg - len(mtext)
print(bogus_req)
bogus = []
bogus_char = Modulus
i = 0
while i < bogus_req:
    bogus_char = bogus_char - 1
    bogus.append(PT[bogus_char])
    i+=1
print(bogus)
```

```python
new_mtext = mtext + bogus
print(new_mtext)
```

```python
num_block = int(len_msg / len(EK))
print(num_block)
```

```python
block_mtext = []
i = 1
j = 1
while i < num_block + 1:
    up_mtext = []
    while j < len(EK) * i + 1:
        up_mtext.append(new_mtext[j-1])
        j+=1
    block_mtext.append(up_mtext)
    i+=1
print(block_mtext)
```

```python
cytext = []
mrg_cytext = []
i = 0
while i < num_block:
    block_cytext = []
    j = 0
    while j < len(EK):
        block_cytext.append(block_mtext[i][int(EK[j]) - 1])
        j+=1
    cytext.append(block_cytext)
    mrg_cytext.append(' '.join(cytext[i]).replace(' ', ''))
    i+=1
# print(cytext)

mrg_cytext = ' '.join(mrg_cytext).replace(' ', '')
print(f'CipherText: {mrg_cytext}')
```

<h3>Decrypting Message</h3>

```python
# CypherText
rcv_cytext = mrg_cytext      # rcv_cytext is received ciphertext
print(f'CipherText: {rcv_cytext}')
```

```python
rcv_cytext = list(rcv_cytext)
print(rcv_cytext)
num_block = int(len(rcv_cytext) / len(EK))
print(num_block)
```

```python
# Generating numbers in sequence till the length of EK
num_seq = []     # num_seq if numbers in sequence
i = 1
while i < len(EK) + 1:
    num_seq.append(str(i))
    i+=1
    
print(f'Block sequence: {num_seq}')
```

<p><b>Note:</b> This block sequence is the position of the encryption key value in sequence. For example:</p>
<pre>
    Block sequence: 1 2 3 4 5 6
    Encryption Key: 2 5 4 3 6 1
</pre>
<p>By rearranging the block sequence based on the rearrangement of the encryption key in accending order we will get decryption key.</p>
<pre>
    Example:
    Block sequence: 6 1 4 3 2 5
    Encryption Key: 1 2 3 4 5 6
</pre>
<p>Here, the rearranged block sequence is the decryption key.</p>

```python
print(f'Encryption Key: {EK}')
```

```python
# Generating decryption key
DK = []
for x in num_seq:
    i = 0
    while i < len(EK):
        if x == EK[i]:
            DK.append(num_seq[i])
        i+=1
print(f'Decryption Key: {DK}')
```

```python
block_cytext = []
i = 1
j = 1
while i < num_block + 1:
    up_cytext = []
    while j < len(DK) * i + 1:
        up_cytext.append(rcv_cytext[j-1])
        j+=1
    block_cytext.append(up_cytext)
    i+=1
print(block_cytext)
```

```python
platext = []
mrg_platext = []
i = 0
while i < num_block:
    block_platext = []
    j = 0
    while j < len(EK):
        block_platext.append(block_cytext[i][int(DK[j]) - 1])
        j+=1
    platext.append(block_platext)
    mrg_platext.append(' '.join(platext[i]).replace(' ', ''))
    i+=1

mrg_platext = ' '.join(mrg_platext).replace(' ', '')
print(f'PlainText: {mrg_platext}')
```
