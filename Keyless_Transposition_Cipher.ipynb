<h1 style = 'text-align: center'><b>Keyless Transposition Cipher</b></h1>
<h3>
    <p>
        The keyless ciphers permute the characters by using writing plaintext in one way (e.g. row by row) and reading it in another way (e.g. column by column). The permutation is done on the whole plaintext to create the whole ciphertext.
        <br>
        <br>
        There are two methods for permutation of characters:
        <ul>
            <li>In the first method, the text is written into a table column by column and then transmitted row by row.</li>
            <li>In the second method, the text is written into a table row by row and then transmitted column by column.</li>
        </ul>
    </p>
    <p><b>***Note: </b> In this code, we use the second method.</p>
</h3>

<h3>1. Generating necessary list of characters</h3>

```python
# Generating Small letters (English alphabets 'a-z')
SL = []
for x in range(ord('a'),ord('z')+1):
    SL.append(chr(x))
# print(SL)

# Generating Capital letters (English alphabets 'A-Z')
CL = []
for x in range(ord('A'),ord('Z')+1):
    CL.append(chr(x))
# print(CL)

# Plain Text Character list
PT = SL
# Cipher Text Character list
CT = CL
print(f'PlainText Characters: {PT}')
print(f'CipherText Characters: {CT}')
Modulus = len(CT)
```

<h3>2. Input Message</h3>

```python
# Input the message
message = input("Write your message here: ")
## Example
# message = "Today is Friday"
print(f'Message: {message}')
```

<h3>3. Encrypting Message</h3>

```python
# List of input characters
inlist = list(message)
is_inlist = 0
while is_inlist<len(inlist):
    if inlist[is_inlist] in CT:
        inlist[is_inlist] = inlist[is_inlist].lower()
    is_inlist+=1
print(inlist)
```

```python
# Removing blank spaces from the message 
mtext = []      # mtext is the message text
for x in inlist:
    if x in PT:
        mtext.append(x)
print(f'The message: {mtext}')
len_msg = len(mtext)
print(f'The length of message {len_msg}.')
```

<p> <b>***Note:</b>
    <ol>
        <li>The message length needs to be greater than 2.</li>
        <li>The length can not be a prime number</li>
    </ol>
</p>

```python
# Function to check whether the number is a prime number
def isPrime(num):
    # Assume number is prime
    is_prime = True

    # If the input is negetive
    if num <= 0:
        is_prime = "Error! Input a positive integer number."
        
    elif num > 1:
        for x in range(2, num):            
            if num % x == 0:
                is_prime = False
    else:
        is_prime = False

    for x in range(2, num):            
        if num % x == 0:
            is_prime = False
            
    return is_prime
```

```python
# Find the factors of a number
def factors(num):
    factors = []
    for x in range(1, num+1):
        if num % x == 0:
            factors.append(x)
    # print(factors)
    return factors
```

```python
# If length of the message is a prime number, then the length will be increased by 1.
# len_msg = 1
is_msglen_prime = isPrime(len_msg)
    
while is_msglen_prime:
    len_msg = len_msg + 1
    is_msglen_prime = isPrime(len_msg)
print(len_msg)
```

```python
# Finding the required number of bogus character & generating bogus character
bogus_req = len_msg - len(mtext)        # bogus_req is the total number of required bogus characters
print(f'Bogus character required: {bogus_req}')
bogus = []
bogus_char = Modulus       # bogus_char is the bogus character value
i = 0
while i < bogus_req:
    bogus_char = bogus_char - 1
    bogus.append(PT[bogus_char])
    i+=1
bogus = bogus[::-1]
print(f'Bogus character list: {bogus}')
```

```python
# Adding the bogus character with the message
new_mtext = mtext + bogus
print(f'The updated mesasge: {new_mtext}')
```

```python
# The list of factor exist of a number
factor_list = factors(len_msg)
print(f'The factors of {len_msg} are {factor_list}.')

# The length of the factor
len_fac = len(factor_list)

# Generating the pair from factors to 
fac_col = 2 
fac_row = len_fac / fac_col
fac_mat = []

i = 1
while i < fac_row + 1:
    up_fac_mat = []
    j = 0
    while j < fac_col:
        if j == 0:
            up_fac_mat.append(factor_list[i-1])
        elif j == 1:
            up_fac_mat.append(factor_list[len_fac - i])
        j+=1
    fac_mat.append(up_fac_mat)
    i+=1
print(f'The pair of factors: {fac_mat}')
```

```python
# Finding the required number of the rows of the m*n matrix
num_row = fac_mat[len(fac_mat)-1][0]
num_col = fac_mat[len(fac_mat)-1][1]
print(f'The required number of rows: {num_row}')
print(f'The required number of columns: {num_col}')
```

```python
# print(type(num_row))
```

```python
# Creating the matrix of the message
block_mtext = []
i = 1
j = 1
while i < num_row + 1:
    up_mtext = []      # up_mtext represent the each block of the mtext
    while j < num_col * i + 1:
        up_mtext.append(new_mtext[j-1])
        j+=1
    block_mtext.append(up_mtext)
    i+=1
print(f'The message matrix: {block_mtext}')
```

```python
# Generating CipherText
col_cytext = []     # col_cytext is the column-wise ciphertext
mrg_cytext = []     # mrg_cytext is the string representation of the ciphertext
i = 0
while i < num_col:
    up_cytext = []
    j = 0
    while j < num_row:
        up_cytext.append(block_mtext[j][i])
        j+=1
    col_cytext.append(up_cytext)
    mrg_cytext.append(' '.join(col_cytext[i]).replace(' ', ''))
    i+=1
mrg_cytext = ' '.join(mrg_cytext).replace(' ', '')
# print(f'Col-wise CipherText: {col_cytext}')
print(f'CipherText: {mrg_cytext}')
```

<h3>4. Decrypting Message</h3>

```python
# CypherText
rcv_cytext = mrg_cytext     # rcv_cytext is received ciphertext
print(f'CipherText: {rcv_cytext}')

# Representing in list
rcv_cytext = list(rcv_cytext)
print(f'CipherText list: {rcv_cytext}')
```

```python
# Number of row and column of the matrix
print(f'Number of rows: {num_row}')
print(f'Number of columns: {num_col}')

# Generating m*n matrix from the message
block_rcv_cytext = []
i = 1
j = 1
while i < num_col + 1:
    up_cytext = []
    while j < num_row * i + 1:
        up_cytext.append(rcv_cytext[j-1])
        j+=1
    block_rcv_cytext.append(up_cytext)
    i+=1
print(f'CipherText Matrix: {block_rcv_cytext}')
```

```python
# Generating PlainText
platext = []
mrg_platext = []
i = 0
while i < num_row:
    block_platext = []
    j = 0
    while j < num_col:
        block_platext.append(block_rcv_cytext[j][i])
        j+=1
    platext.append(block_platext)
    mrg_platext.append(' '.join(platext[i]).replace(' ', ''))
    i+=1

mrg_platext = ' '.join(mrg_platext).replace(' ', '')
print(f'PlainText: {mrg_platext}')
```
